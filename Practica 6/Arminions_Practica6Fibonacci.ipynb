### Introducción y Metodología 

El algoritmo de Fibonacci con enfoque "top-down" y programación dinámica (DP) es una técnica que utiliza la recursión para calcular los números de Fibonacci de manera eficiente, evitando cálculos repetitivos a través de la memorización de los resultados previamente calculados.

El enfoque "top-down" se basa en dividir el problema en subproblemas más pequeños y resolverlos recursivamente. En el caso de Fibonacci, cada número se calcula sumando los dos números anteriores. Sin embargo, en lugar de calcular los números repetidamente, utilizamos la memorización para almacenar los resultados ya calculados y reutilizarlos cuando sea necesario.

El enfoque "bottom-up" en programación dinámica (DP) para el algoritmo de Fibonacci implica resolver los subproblemas más pequeños primero y luego construir gradualmente las soluciones para los subproblemas más grandes. A diferencia del enfoque "top-down" que utiliza la recursión, el enfoque "bottom-up" utiliza iteraciones y un arreglo o diccionario para almacenar los resultados de los subproblemas.

```python
import matplotlib.pyplot as plt
import timeit

def fibo_topdown(n, memo={}):
    if n in memo:
        return memo[n]
    
    if n <= 2:
        result = 1
    else:
        result = fibo_topdown(n-1, memo) + fibo_topdown(n-2, memo)
    
    memo[n] = result
    return result


def fibo_bottomup(n):
    dp = [0] * (n+1)
    dp[0] = 0
    dp[1] = 1
    
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    
    return dp[n]

tamaños = [15, 28, 40, 77]
top_down_tiempos = []
bottom_up_tiempos = []
for tamaño in tamaños:
    top_down_tiempo = timeit.timeit(f'fibo_topdown({tamaño})', globals=globals(), number=1)
    bottom_up_tiempo = timeit.timeit(f'fibo_bottomup({tamaño})', globals=globals(), number=1)
    top_down_tiempos.append(top_down_tiempo)
    bottom_up_tiempos.append(bottom_up_tiempo)
    print(f"Tiempo de ejecución (Top-Down) para tamaño {tamaño}: {top_down_tiempo}")
    print(f"Tiempo de ejecución (Bottom-Up) para tamaño {tamaño}: {bottom_up_tiempo}")
    print("----------------------------------------------------------------------")

plt.plot(tamaños, top_down_tiempos, label='Top-Down')
plt.plot(tamaños, bottom_up_tiempos, label='Bottom-Up')
plt.xlabel('Tamaño')
plt.ylabel('Tiempo (segundos)')
plt.legend()
plt.show()
```

## Complejidad temporal y espacial

En el enfoque "top-down", se utiliza recursión para calcular los números de Fibonacci, pero con la memoización para evitar cálculos repetitivos. A medida que se resuelven los subproblemas, los resultados se almacenan en un diccionario o en una tabla para su reutilización. La complejidad temporal sigue siendo O(n) ya que, en el peor de los casos, se realiza un cálculo para cada número de Fibonacci desde 2 hasta n. La complejidad espacial también es O(n) debido al uso del diccionario o la tabla para almacenar los resultados de los subproblemas.

En el enfoque "bottom-up", se iteran los números de Fibonacci desde 2 hasta n, calculando cada número en función de los dos números anteriores y almacenándolos en un arreglo. Como resultado, el tiempo de ejecución crece linealmente con respecto a n, lo que da una complejidad temporal de O(n). Además, se utiliza un arreglo para almacenar los resultados de los subproblemas, lo que resulta en una complejidad espacial de O(n), ya que se requiere un arreglo de tamaño n+1 para almacenar todos los valores de Fibonacci hasta el número deseado.

## Conclusiones

Ambos enfoques tienen una complejidad temporal lineal y una complejidad espacial lineal, lo que los hace eficientes para calcular números de Fibonacci. La principal diferencia radica en cómo se aborda el cálculo: el enfoque "top-down" utiliza recursión y memoización, mientras que el enfoque "bottom-up" utiliza iteración y almacenamiento en una tabla.

En general, se puede concluir que ambos enfoques programados son eficientes para los tamaños dados.
